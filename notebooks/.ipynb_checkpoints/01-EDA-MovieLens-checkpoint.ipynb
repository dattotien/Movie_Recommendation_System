```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MovieLens EDA") \
    .master("local[*]") \
    .config("spark.driver.bindAddress", "127.0.0.1") \
    .getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, FloatType

# Định nghĩa schema cho ratings.dat
ratings_schema = StructType([
    StructField("UserID", IntegerType(), True),
    StructField("MovieID", IntegerType(), True),
    StructField("Rating", FloatType(), True),
    StructField("Timestamp", IntegerType(), True)
])
from pyspark.sql.types import StringType

# users.dat
users_schema = StructType([
    StructField("UserID", IntegerType(), True),
    StructField("Gender", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Occupation", IntegerType(), True),
    StructField("Zipcode", StringType(), True)
])

# movies.dat
movies_schema = StructType([
    StructField("MovieID", IntegerType(), True),
    StructField("Title", StringType(), True),
    StructField("Genres", StringType(), True)
])

```

```python

# Đọc ratings.dat
ratings = spark.read.csv(
    "../data/ml-1m/ratings.dat",
    sep="::",
    schema=ratings_schema,
    header=False
)

# Đọc users.dat
users = spark.read.csv(
    "../data/ml-1m/users.dat",
    sep="::",
    schema=users_schema,
    header=False
)

# Đọc movies.dat
movies = spark.read.csv(
    "../data/ml-1m/movies.dat",
    sep="::",
    schema=movies_schema,
    header=False
)
```

```python
# Số lượng bản ghi
print(f"Số lượng ratings: {ratings.count()}")
print(f"Số lượng users: {users.count()}")
print(f"Số lượng movies: {movies.count()}")

# Rating min, max, mean
ratings.describe("Rating").show()

# Số lượng rating khác nhau
ratings.groupBy("Rating").count().orderBy("Rating").show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Chuyển về pandas DataFrame (chỉ lấy Rating)
ratings_pd = ratings.select("Rating").toPandas()

# Vẽ histogram phân bố rating
plt.figure(figsize=(6,4))
plt.hist(ratings_pd["Rating"], bins=[0.5,1.5,2.5,3.5,4.5,5.5], edgecolor='black')
plt.title("Phân bố điểm đánh giá (Ratings)")
plt.xlabel("Điểm (1-5)")
plt.ylabel("Số lượng")
plt.show()
```

```python
# Số lượng rating mỗi user
user_activity = ratings.groupBy("UserID").count().withColumnRenamed("count", "NumRatings")

# Số lượng rating mỗi movie
movie_popularity = ratings.groupBy("MovieID").count().withColumnRenamed("count", "NumRatings")

# Lấy về pandas để vẽ
user_activity_pd = user_activity.toPandas()
movie_popularity_pd = movie_popularity.toPandas()

# Biểu đồ phân phối số lượng rating theo người dùng
plt.figure(figsize=(6,4))
plt.hist(user_activity_pd["NumRatings"], bins=50, color='skyblue', edgecolor='black')
plt.title("Phân phối số lượng rating theo người dùng")
plt.xlabel("Số lượng rating")
plt.ylabel("Số người dùng")
plt.show()

# Biểu đồ phân phối số lượng rating theo phim
plt.figure(figsize=(6,4))
plt.hist(movie_popularity_pd["NumRatings"], bins=50, color='salmon', edgecolor='black')
plt.title("Phân phối số lượng rating theo phim")
plt.xlabel("Số lượng rating")
plt.ylabel("Số phim")
plt.show()
```

```python
from pyspark.sql.functions import desc

# Join để có tên phim
top_movies = movie_popularity.join(movies, "MovieID") \
    .orderBy(desc("NumRatings")) \
    .select("Title", "NumRatings")

top_movies.show(10, truncate=False)
```

```python
from pyspark.sql.functions import avg

# Join users và ratings
user_ratings = ratings.join(users, "UserID")

# Trung bình rating theo giới tính
avg_rating_gender = user_ratings.groupBy("Gender").agg(avg("Rating").alias("AvgRating"))
avg_rating_gender.show()

# Trung bình rating theo độ tuổi
avg_rating_age = user_ratings.groupBy("Age").agg(avg("Rating").alias("AvgRating"))
avg_rating_age.orderBy("Age").show()
```

```python
avg_rating_gender_pd = avg_rating_gender.toPandas()

plt.bar(avg_rating_gender_pd["Gender"], avg_rating_gender_pd["AvgRating"], color=['pink','skyblue'])
plt.title("Trung bình điểm đánh giá theo giới tính")
plt.xlabel("Giới tính")
plt.ylabel("Điểm trung bình")
plt.show()
```

```python
from pyspark.sql.functions import from_unixtime

ratings = ratings.withColumn("Datetime", from_unixtime("Timestamp"))
```

```python
from pyspark.sql.functions import split

movies = movies.withColumn("GenresArray", split("Genres", r"\|"))
```

```python
# Join ratings với users và movies
full_df = ratings.join(users, on="UserID").join(movies, on="MovieID")
full_df.show(5)
```

```python
# Example: lọc rating ngoài 1-5
full_df = full_df.filter((full_df.Rating >= 1) & (full_df.Rating <= 5))
```

```python
from pyspark.sql.functions import col

# Chuẩn hóa cho ALS
als_data = ratings.select(
    col("UserID").alias("userId"),
    col("MovieID").alias("movieId"),
    col("Rating").cast("float")
)

# Drop các hàng null
als_data = als_data.dropna()

# Kiểm tra số lượng records
print("Records ready for ALS:", als_data.count())
```
